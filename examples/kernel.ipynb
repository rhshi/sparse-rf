```python
import numpy as np 
import matplotlib.pyplot as plt 
from sparse_rf.algs.core import min_wl2, min_l2, min_l1
from sparse_rf.modules import make_X, make_W, make_A, make_K
from sparse_rf.util import *
from sparse_rf.util.fns import *
from functools import partial
import random
import gc
```

```python
m = 1000
d = 5
active = fourier

ratio_train = 0.7
Ns = np.arange(1, 10201, 200)
```

```python
X = make_X(d, m, dist=uniform)
X_train = X[:int(m*ratio_train), :]
X_test = X[int(m*ratio_train):, :]

y = np.array(list(map(fn2, X)))
y_train = y[:int(m*ratio_train)]
y_test = y[int(m*ratio_train):]
```

```python
q = 1

K_results = []
A_results = []

K_train = make_K(X_train, kernel=partial(sparse_rbf, d=d, q=q))
K_test = make_K(X_train, X_test=X_test, kernel=partial(sparse_rbf, d=d, q=q))
c_K = min_l2(K_train, y_train)

K_results.append(np.linalg.norm(K_test@c_K-y_test) / np.linalg.norm(y_test))

for N in Ns:
    A_res = []
    for _ in range(1):
        W = make_W(d, q, N=N, dist=normal)
        A_train = make_A(X_train, W, active=active) / np.sqrt(N)
        A_test = make_A(X_test, W, active=active) / np.sqrt(N)
        c_A = min_l2(A_train, y_train)
        A_res.append(np.linalg.norm(A_test@c_A-y_test) / np.linalg.norm(y_test))

        del W
        del A_train
        del A_test
        del c_A
        gc.collect()

    A_results.append(A_res)

del K_train
del K_test
del c_K 
gc.collect()

plt.semilogy(2*Ns, [np.mean(res) for res in A_results], '*-')
plt.semilogy(2*Ns, [K_results[0]] * len(Ns), "--")
plt.show()
```
